```python
import os
import shutil
import random
from pathlib import Path

# Configuration
source_dir = Path("/home/student/sebastian_ma/VoronoiLoss/data/WMH_raw")
output_dir = Path("/home/student/sebastian_ma/VoronoiLoss/data/brain/wmh")
splits = {'train': 0.7, 'val': 0.15, 'test': 0.15}

# 1. Gather all case IDs (subdirectories in 'base')
case_ids = [d.name for d in source_dir.iterdir() if d.is_dir()]
random.seed(42)  # For reproducibility
random.shuffle(case_ids)

# 2. Calculate split indices
total = len(case_ids)
train_idx = int(total * splits['train'])
val_idx = train_idx + int(total * splits['val'])

split_map = {
    'train': case_ids[:train_idx],
    'val': case_ids[train_idx:val_idx],
    'test': case_ids[val_idx:]
}
print(split_map)

# 3. Create folders and move files
for split_name, ids in split_map.items():
    img_path = output_dir / split_name / 'images'
    lbl_path = output_dir / split_name / 'labels'
    
    img_path.mkdir(parents=True, exist_ok=True)
    lbl_path.mkdir(parents=True, exist_ok=True)

    for cid in ids:
        src_img = source_dir / cid / 'pre' / 'T1.nii.gz'
        src_lbl = source_dir / cid / 'wmh.nii.gz'
        os.makedirs(f"{img_path}/WMH_{cid}", exist_ok=True)
        
        # Rename files to include case ID to avoid name collisions
        shutil.copy2(src_img, img_path / f"WMH_{cid}/T1.nii.gz")
        shutil.copy2(src_lbl, lbl_path / f"WMH_{cid}.nii.gz")

print(f"Reorganization complete! Total cases: {total}")
```
